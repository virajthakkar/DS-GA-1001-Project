```python
import math
import matplotlib.pylab as plt

from datetime import datetime as dt

#adjust to each CCY
df = pd.read_csv('All MXN Data.csv')
```

```python
df['Trend Start Time'] = pd.to_datetime(df['Trend Start Time'])
df['Trend Start Time'] = df['Trend Start Time'].dt.hour * 60 + df['Trend Start Time'].dt.minute
df
```

```python
#df = df.drop(['Day Start'],axis = 1)
df = df.dropna()
# df = df.drop([''], axis=1)
df['Trend Type 2']  = np.where(df['Trend Type'] == 'Up', 1, 0)
df = df.drop('Trend Type', axis=1)
df = df.reset_index()
```

```python
df
```

```python
from sklearn import preprocessing
from sklearn import neighbors
from sklearn.model_selection import train_test_split


dfK = df[['USDMXN ROC 4', 'USDMXN MACD', 'MXEF Tick', 'Trend Start Time', 'BBDXY Change', 'Previous Day DXY Trend',
         'Previous Day SPY Trend', 'V1M Change']]
#dfK_y = df[['Target']]


x = dfK.values #returns a numpy array
min_max_scaler = preprocessing.MinMaxScaler()
x_scaled = min_max_scaler.fit_transform(x)
dfK = pd.DataFrame(x_scaled)
dfK['Target'] = df['Target']

X_train, X_test, y_train, y_test = train_test_split(df.drop('Target', axis=1), df['Target'], test_size=0.33, random_state=45)

knn = neighbors.KNeighborsClassifier(n_neighbors=3)

knn.fit(X_train,y_train)

probs = pd.DataFrame(knn.predict_proba(X_test))

probs['Prediction'] = ""

for i in range(len(probs)):
    if probs.loc[i,0] < .2:
        probs.loc[i,'Prediction'] = 1
    elif probs.loc[i,0] < .4:
        probs.loc[i,'Prediction'] = 2
    elif probs.loc[i,0] < .5:
        probs.loc[i,'Prediction'] = 3
    elif probs.loc[i,0] < .6:
        probs.loc[i,'Prediction'] = 4
    elif probs.loc[i,0] < .8:
        probs.loc[i,'Prediction'] = 5
    elif probs.loc[i,0] <= 1:
        probs.loc[i,'Prediction'] = 6
probs
```

```python
from sklearn import metrics
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
fpr_knn, tpr_knn, thresholds_lr = metrics.roc_curve(y_test, knn.predict_proba(X_test)[:,1])
roc_auc_knn = metrics.auc(fpr_knn, tpr_knn)

# Code here
plt.figure(figsize = (10,10))
plt.plot(fpr_knn, tpr_knn, label = 'KNN (AUC = %0.3f)' % roc_auc_knn)
plt.legend(loc="lower right")
plt.show()
```

```python
from sklearn.decomposition import PCA
from sklearn import preprocessing as pp

df_copy = df.copy()

X_train, X_test, y_train, y_test = train_test_split(df_copy.drop('Target', axis=1), df_copy['Target'], test_size=0.33, random_state=42)

# df_copy[l1] = pd.DataFrame(pp.StandardScaler(with_std=False).fit_transform(df_copy[l1]))
train_mean = X_train.mean()

X_train=(X_train-train_mean)

X_test = (X_test-train_mean)

pca = PCA(n_components=2)
pca.fit(X_train)
print(pca.explained_variance_ratio_)

k=2

if k is None:
    k = len(pca.explained_variance_ratio_)

sum1 = np.sum(pca.explained_variance_ratio_)
sum2 = 0
attachArray = []
for i in range(k):
    sum2 = sum2 + pca.explained_variance_ratio_[i]
    attachArray.append(sum2/sum1)

x = np.arange(1,len(pca.explained_variance_ratio_[:k])+1)
y = np.array(attachArray[:k]) * 100

plt.figure(figsize=(5, 5))

plt.plot(x,y)
plt.xlabel('i')
plt.ylabel('Percentage variance covered by first \'i\' eigenvalues')
plt.title('Variance Covered')
plt.show()
plt.close()
```

```python
principalComponents = pca.fit_transform(X_train)
principalDf = pd.DataFrame(data = principalComponents)

clf = RandomForestClassifier(n_estimators=1000, max_depth=5,
                             random_state=0)
clf.fit(principalDf, y_train)
pricipalDf_X_test = pca.transform(X_test)
y_pred = clf.predict(pricipalDf_X_test)

accuracy = accuracy_score(y_pred, y_test)
print(accuracy)
```
